```python
from vtkplotter import *
import numpy as np

embedWindow('2d')
np.random.seed(3)

##################################################################
radhisto = histogram(np.random.rand(200)*6.28,
                     mode='polar',
                     title="random orientations",
                     bins=10,
                     c='orange', #uniform color
                     labels=["label"+str(i) for i in range(10)],
                    )

show(radhisto)
```
