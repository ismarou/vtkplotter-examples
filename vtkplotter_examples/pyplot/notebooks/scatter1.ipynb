```python
"""A simple scatter plot with plotxy()"""
from vtkplotter import *
from numpy.random import randn
embedWindow('2d')

x = randn(100)
y = randn(100)*20

plt = plot(
            x, y,       # accepts different formats
            xtitle="variable x",
            ytitle="variable y",
            line=False,
            marker=".",   # marker style
            mc="dr",      # marker color
)

show(plt, axes=1, zoom=1.4)
```
