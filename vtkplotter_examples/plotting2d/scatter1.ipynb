```python
"""A simple scatter plot with plotxy()"""
from vtkplotter import *
from numpy.random import randn
embedWindow('2d')

x = randn(100)
y = randn(100)*20

plt = plotxy(
    [x, y],       # accepts different formats
    xtitle="variable x",
    ytitle="variable y",
    line=False,
    yscale=None,  # automatic y scale
    marker=".",   # marker style
    mc="dr",      # marker color
)

vp = Plotter(bg="w", axes=1)
show(plt, zoom=1.4)
```
